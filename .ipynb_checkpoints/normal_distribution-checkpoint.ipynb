# 正态分布曲线绘制

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
plt.plot(np.linspace(-4,4,100),stats.norm.pdf(np.linspace(-4,4,100))) # 从（-4，4）中随机选取100个数，绘制该事件的概率密度函数
plt.fill_between(np.linspace(-4,4,100),stats.norm.pdf(np.linspace(-4,4,100)),alpha=.15); # 对曲线内部进行填充
```

# 随机生成正态分布案例

```python
mu, sigma = 0.5, 0.5
lower, upper = 0, 1  # 截断在[μ-2σ, μ+2σ]
X = stats.truncnorm((lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
N = stats.norm(loc=mu, scale=sigma)
# plt.hist(X.rvs(10000), bins=30, alpha=0.3);   # 截断正态分布的直方图

plt.figure(1)
plt.subplot(2,1,1)
plt.hist(X.rvs(10000), bins=30, alpha=0.5)   # 截断正态分布的直方图
plt.subplot(2,1,2)
plt.hist(N.rvs(10000), bins=30, alpha=0.5)   # 常规正态分布的直方图
plt.show()
```

# 绘制正态分布曲线

- [曲线绘制](https://www.cnblogs.com/pyhou/p/12535208.html)  
- [标签位置设定](https://zhuanlan.zhihu.com/p/99531531)

'upper left' , 'upper center', 'upper right',  
'center left', 'center', 'center right',  
'lower left', 'lower center', 'lower right'

```python
def normfun(x, mu, sigma):
    pdf = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))
    return pdf
```

```python
# RPR Section 3.2 
# Figure 1
x = np.arange(0, 1, 0.001) 
mean_list = [0.2, 0.5, 0.8] 
std = 0.05 

ax = plt.axes()
for mean in mean_list:
    y = normfun(x, mean, std) / 50
    ax.plot(x, y, label='$\mu={}\%,\sigma^2$={}'.format(int(mean*100), std))
    # ax.fill_between(x, y, alpha=.15)

# 去除y轴刻度
# ax.yaxis.set_major_locator(plt.NullLocator())  

# 设置坐标轴为百分比
def to_percent(temp, position):
    return '%.0f'%(100 * temp) + '%'
from matplotlib.ticker import FuncFormatter
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_percent))

# 打开标签
font_label = {'size': 11}  # 标签字体
plt.legend(prop=font_label, # 标签字体
           loc='upper right' # 标签位置
          )  
# lower 
plt.xlim(0, 1)
# plt.ylim(0, 1)
plt.xticks(fontsize= 13)
plt.yticks(fontsize= 13)

plt.xlabel('Probe Vehicle RPR', fontdict={'weight': 'normal', 'size': 13})
plt.ylabel('Probability Density Distribution', fontdict={'weight': 'normal', 'size': 13});

# 需要避免图片不完整
plt.savefig('data_distribution/rpr_distribution_3-2_fig1.png', dpi=300, bbox_inches = 'tight')
```

```python
# section 3.2 Figure 3
distribution_list = [(0.2, 0.05), (0.5, 0.05), (0.8, 0.05)]
# mu, sigma = 0.2, 0.05

ax = plt.axes()
for mu, sigma in distribution_list:
    lower, upper = 0, 1  # 截断在[μ-2σ, μ+2σ]
    X = stats.truncnorm((lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
    # plt.hist(X.rvs(10000), bins=30, alpha=0.3);   # 截断正态分布的直方图
    ax.hist(X.rvs(50), bins=30, alpha=0.5, label='$\mu={}\%,\sigma^2$={}'.format(int(100 * mu), sigma))   # 截断正态分布的直方图

# # 设置坐标轴为百分比
def to_percent(temp, position):
    return '%.0f'%(100 * temp) + '%'
from matplotlib.ticker import FuncFormatter
# plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_percent))

# 打开标签
font_label = {'size': 11}  # 标签字体
plt.legend(prop=font_label, # 标签字体
           loc='upper right' # 标签位置
          )  
# lower 
plt.xlim(0, 1)
plt.ylim(0, 8)
plt.xticks(fontsize= 13)
plt.yticks(fontsize= 13)

plt.xlabel('Probe Vehicle RPR', fontdict={'weight': 'normal', 'size': 13})
plt.ylabel('Number of Routes', fontdict={'weight': 'normal', 'size': 13});

# 需要避免图片不完整
plt.savefig('data_distribution/rpr_ramdom_sample_3-2_fig3.png', dpi=300, bbox_inches = 'tight')
```

```python
# Exp 1
# 步长设定得越小，画出来的正态分布曲线越平滑
x = np.arange(0, 1, 0.005)
mean = 0.1 
std = 0.1 
# 设定Y轴，载入刚才定义的正态分布函数
# y = normfun(x, mean, std)
# print(y)

# 绘制数据集的正态分布曲线
ax = plt.axes()
# std_list = [0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50]
std_list = [0.05, 0.10, 0.15, 0.50]
for std in std_list:
    y = normfun(x, mean, std) / 50
    ax.plot(x, y, label='$\sigma^2$={}'.format(std))
    # ax.fill_between(x, y, alpha=.15)

# ax.yaxis.set_major_locator(plt.NullLocator())  # 去除y轴刻度
# 设置坐标轴为百分比
def to_percent(temp, position):
    return '%.0f'%(100 * temp) + '%'
from matplotlib.ticker import FuncFormatter
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_percent))


font_label = {'size': 13}  # 标签字体
plt.legend(prop=font_label)  # 打开标签
plt.xlim(0, 1)
plt.xticks(fontsize= 13)
plt.yticks(fontsize= 13)
plt.xlabel('Probe Vehicle RPR ($\gamma_c=10\%$)', fontdict={'weight': 'normal', 'size': 13})
plt.ylabel('Probability Density Distribution', fontdict={'weight': 'normal', 'size': 13});
plt.savefig('data_distribution/rpr_distribution_exp_1.png', dpi=300, bbox_inches = 'tight')
```

```python
# Exp 2

# 步长设定得越小，画出来的正态分布曲线越平滑
x = np.arange(0, 1, 0.005)
mean = 0.1
std = 0.2
# 设定Y轴，载入刚才定义的正态分布函数
# y = normfun(x, mean, std)
# print(y)

# 绘制数据集的正态分布曲线
ax = plt.axes()
mean_list = [round(i*0.05, 2) for i in range(1, 21, 2)]
for mean in mean_list:
    y = normfun(x, mean, std) / 50
    ax.plot(x, y, label='$\gamma_c$={}%'.format(int(mean*100)))
    # ax.fill_between(x, y, alpha=.15)
    
# ax.yaxis.set_major_locator(plt.NullLocator())  # 去除y轴刻度
# 设置坐标轴为百分比
def to_percent_1(temp, position):
    return '%.0f'%(100 * temp) + '%'
def to_percent_2(temp, position):
    return '%.1f'%(100 * temp) + '%'
from matplotlib.ticker import FuncFormatter
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_percent_1))
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent_2))


font_label = {'size': 11}  # 标签字体
plt.legend(prop=font_label)  # 打开标签
plt.xlim(0, 1)
plt.xticks(fontsize= 13)
plt.yticks(fontsize= 13)
plt.xlabel('Probe Vehicle RPR ($\sigma^2$=0.20)', fontdict={'weight': 'normal', 'size': 13})
plt.ylabel('Probability Density Distribution', fontdict={'weight': 'normal', 'size': 13});
plt.savefig('data_distribution/rpr_distribution_exp_2.png', dpi=300, bbox_inches = 'tight')
```
